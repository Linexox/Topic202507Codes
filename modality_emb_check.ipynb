```python
ado_folder = "datasets\\DuRecDial\\audio_emb"
img_folder = "datasets\\DuRecDial\\img_emb"
txt_folder = "datasets\\DuRecDial\\txt_emb"
vdo_folder = "datasets\\DuRecDial\\vdo_emb"
```

```python
import os
import numpy as np
import torch

ado_folder2 = "datasets\\DuRecDial\\ado_emb"


def compare_audio_vectors(folder1, folder2):
    # Load audio vectors from .pt files
    vectors1 = {
        f: torch.load(os.path.join(folder1, f)) for f in os.listdir(folder1) if f.endswith(".pt")
    }
    vectors2 = {
        f: torch.load(os.path.join(folder2, f)) for f in os.listdir(folder2) if f.endswith(".pt")
    }

    common_files = set(vectors1.keys()).intersection(set(vectors2.keys()))
    results = {}

    for f in common_files:
        results[f] = np.array_equal(vectors1[f], vectors2[f])

    return results


comparison_results = compare_audio_vectors(ado_folder, ado_folder2)
print(comparison_results)
```

```python
def compare_audio_shapes(folder1, folder2):
    shapes1 = check_embeddings_shape(folder1)
    shapes2 = check_embeddings_shape(folder2)

    comparison = {}
    for f in set(shapes1.keys()).union(set(shapes2.keys())):
        shape1 = shapes1.get(f, "File not found")
        shape2 = shapes2.get(f, "File not found")
        comparison[f] = (shape1, shape2)

    return comparison


audio_shape_comparison = compare_audio_shapes(ado_folder, ado_folder2)
print(audio_shape_comparison)
```

```python
def check_embeddings_shape(folder):
    shapes = {}
    for f in os.listdir(folder):
        if f.endswith(".pt"):
            tensor = torch.load(os.path.join(folder, f))
            shapes[f] = tensor.shape
    return shapes


# 检查每个文件夹的嵌入形状
audio_shapes = check_embeddings_shape(ado_folder)
img_shapes = check_embeddings_shape(img_folder)
txt_shapes = check_embeddings_shape(txt_folder)
vdo_shapes = check_embeddings_shape(vdo_folder)

print("Audio Shapes:", audio_shapes)
print("Image Shapes:", img_shapes)
print("Text Shapes:", txt_shapes)
print("Video Shapes:", vdo_shapes)
```
